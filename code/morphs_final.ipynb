```python
import re
replace = {"갠적" : "개인적", "엉댕이" : "엉덩이", "배이지" : "베이지", "언냐" : "언니", "져아" : "좋아", "조와" : "좋아", 
           "조아" : "좋아", "욥" : "요", "시름" : "싫다", "일부로" : "일부러", "나서" : "나다", "에오" : "에요", "욤" : "요",
          "푠" : "편", "큐션" : "쿠션", "다요" : "아요", "다용" : "아요", "넘무" : "너무", "젛아" : "좋아", "꿰밌" : "꿰멨",
          "갱얼쥐" : "강아지", "고영이" : "고양이", "애옹이" : "고양이", "휘두루마뚜루" : "휘뚜루마뚜루", "쮹" : "쭉", "고냥이" : "고양이",
           "냥이" : "고양이", "쉬야" : "오줌", "멍멍이": "강아지", "멈머" : "강아지", "지듀" : "지도", "나뿌" : "나쁘", "뇽" : "요", 
           "이따구" : "이따위", "이딴" : "이런", "풋신" : "푹신", "냥" : "고양이", "짜투리" : "자투리", "놥뒀" : "놔뒀", "샊갈" : "색깔",
          }
out = []
for i, rev in enumerate(review):
    #print("Before : ", rev)
    for rep in replace.keys():
        if rep in rev:
            rev = re.sub(rep, replace.get(rep), rev)
    out.append(rev)
    print(i, "After : ", rev, "\n")
```
