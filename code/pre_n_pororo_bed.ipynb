```python
# n-gram 키워드 추출 함수 정의
def get_feature_keywords(feature_keywords, review):
    feature_temp = []
    for keyword in feature_keywords:
        if re.findall(keyword, review):
            #가격이 진짜 너무 싸요
            a = re.findall(keyword +'+[ㄱ-ㅎ|ㅏ-ㅣ|가-힣|0-9|a-z|A-Z]+\s+[ㄱ-ㅎ|ㅏ-ㅣ|가-힣|0-9|a-z|A-Z]+\s+[ㄱ-ㅎ|ㅏ-ㅣ|가-힣|0-9|a-z|A-Z]+\W*', review) 
            #가격이 진짜 싸요
            b = re.findall(keyword +'+[ㄱ-ㅎ|ㅏ-ㅣ|가-힣|0-9|a-z|A-Z]+\s+[ㄱ-ㅎ|ㅏ-ㅣ|가-힣|0-9|a-z|A-Z]+\W*', review) 
            #가격 진짜 싸요
            c = re.findall(keyword + '+\s+[ㄱ-ㅎ|ㅏ-ㅣ|가-힣|0-9|a-z|A-Z]+\s+[ㄱ-ㅎ|ㅏ-ㅣ|가-힣|0-9|a-z|A-Z]+', review) 
            #가격이 싸다.(문장 끝)
            d = re.findall('[ㄱ-ㅎ|ㅏ-ㅣ|가-힣|0-9|a-z|A-Z]*'+keyword + '+[ㄱ-ㅎ|ㅏ-ㅣ|가-힣|0-9|a-z|A-Z]+\s+[ㄱ-ㅎ|ㅏ-ㅣ|가-힣|0-9|a-z|A-Z]+\\*$', review)
            #가격 싸다.(문장 끝)
            e = re.findall('[ㄱ-ㅎ|ㅏ-ㅣ|가-힣|0-9|a-z|A-Z]*'+keyword + '+\s+[ㄱ-ㅎ|ㅏ-ㅣ|가-힣|0-9|a-z|A-Z]+\W*$', review) 
            ft = a + b + c + d + e
            if len(ft) != 0:
                ft = list(pd.DataFrame(ft)[0].unique())
            else:
                continue

            #print(keyword, " 오리지널 : ", ft)
            idx_out = []
            for i in range(len(ft)):
                for_last = ft[i].split(" ")
                if len(for_last[-1]) != 0:
                    last = for_last[-1]
                else:
                    last = for_last[-2]
                for j in range(len(feature_keywords)):
                    if feature_keywords[j] in last:
                        idx_out.append(i)
            for idx in sorted(list(np.unique(idx_out)), reverse = True):
                del ft[idx]
            #print("키워드 중복 방지 : ", ft)

            idx_out = []
            ft.sort(key = lambda x: len(x), reverse = True)
            for i in range(len(ft)-1):
                for j in range(i+1, len(ft)):
                    if ft[j] in ft[i]:
                        idx_out.append(j)
            for idx in sorted(list(np.unique(idx_out)), reverse = True):
                del ft[idx]

            #print("중복 제거됨 : ", ft)
            for ngram in ft:
                feature_temp.append(ngram)
                ngram = re.sub("\s*\W+$", "", ngram)
                review = re.sub(ngram + "\W*", "", review)

    print(feature_temp, "\n")
    return feature_temp
```

```python
def mypororo(keywords):
  #키워드 부분 추출 
  res = []
  for i in range(len(df)):
    res.append(get_feature_keywords(keywords, df['review2'][i]))

  #뽀로로 감성 점수
  scores = []
  for r in res:
    r = ' '.join(r)                     #한 리뷰에서 키워드 부분 여러 개면 하나로 합치기 
    scores.append(pororo(r))            #뽀로로 돌리기

  return scores
```
